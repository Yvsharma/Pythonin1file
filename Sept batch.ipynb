# User define error

```python
class myclassError(Exception):
    pass

try:
    raise myclassError
except myclassError:
    print("caught error")
```

```python
class myclassError(Exception):
    pass

try:
    for i in range(1,11,2):
        if i==6:
            raise myclassError("no. 6 can't print")
        print("value from loop is : ",i)
except myclassError:
    print("caught error")
```
